```python
import numpy as np
import matplotlib.pyplot as plt
nums = [200,500,1000,2000,5000,10000,20000]

with open('pa1b/hp.txt','rb') as f:
    table = np.zeros((7,10))
    for l in f.readlines():
        ll = list()
        for i in l.split():
            ll.append(float(i))
        
        # body number
        ind = 0
        num = 0
        for i in range(len(nums)):
            if abs(ll[0]-nums[i])<0.1:
                num = i
        while table[num][ind]!=0:
            ind = ind+1
        #print num, ind
        table[num][ind]=ll[2]

    print table
    
    
    plt.title('performance')
    plt.ylabel('performance k*n*n/t')
    plt.xlabel('amount of bodies')
    plt.xticks(range(len(nums)),nums) 
    # different amount of threads
    for i in range(table.shape[1]):
        # different amount of bodies
        plt.plot(range(table.shape[0]), table[:,i], label=2**(i-1))
    plt.legend()
    plt.show()
    
```
